```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/aadhar_2011.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.drop('Sr.No.',axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.info()
```

```python
df.rename(columns={'State/UT':'State'})
```

```python
x=df[['State/UT','Population as per 2011 census','Aadhaar generated as on 30.11.2016']]
y=df[['% Aadhaar generated']]
```

```python
y.info()
```

```python
# handle categorical variable
states=pd.get_dummies(df,drop_first=True)
# dropping extra column
x= x.drop(x.columns[0],axis=1)
# concatation of independent variables and new cateorical variable.
x=pd.concat([x,states],axis=1)
x
```

```python
x.drop(x.columns[4],axis=1,inplace=True)
```

```python
x.info()
```

```python
y.info()
```

```python
# importing train_test_split from sklearn
from sklearn.model_selection import train_test_split
# splitting the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
# importing module
from sklearn.linear_model import LinearRegression
# creating an object of LinearRegression class
LR = LinearRegression()
# fitting the training data
LR.fit(x_train,y_train)
```

```python
y_prediction =  LR.predict(x_test)
y_prediction
```

```python
y_test=np.array(y_test)
```
